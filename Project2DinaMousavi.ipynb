**Multiple Sequence Alignment:**

```python
#Using biopython, import Bio.AlignIO to align multiple sequences in a single file (one human and 3 Monkey species)
#There are also two human isomer sequences used for reference
from Bio import AlignIO
```

```python
#Define the global variable alignment, which is defined outside of a function
#This is global since this file (psen1.aln) can be assessed through functions such as .read and used in this entire code
#My mac did not let me open a .fasta file, so I converted my sequences to a .aln file which is for Cluster Alignment
#This function will read my file and prepare for the printing phase in the next few steps
alignment = AlignIO.read("psen1.aln", "clustal")
```

```python
#This function is used to find Alignment Length
#This is useful because we want our sequences to be similar in length in order to align properly
print("Alignment length %i" % alignment.get_alignment_length())
```

```python
#Used this function to write out the full sequences from each species
#record.seq refers to a sequence being inputted and read out
#record.id is used to identify the sequence and write it out
#AAH11729.1 for humans, XP_024087009.1 for orangutans, XP_007985388.1 for green monkeys, XP_012293630.1 for night monkey's
#AAB46370.1 and NP_000012.1 are human isoforms for comparison/reference
for record in alignment:
    print(record.seq + " " + record.id)
```

**Part 4: Analysis of the results**

The biggest part of my question was whether this Familial Alziemer's could occur in other species (other than humans) as well. Do Presenilin-1 mutations occuring near the active site of the gamma secretase enzyme affect other species such as monkey's (known to develop Familial Alziemer's) or species that are not known to develop Alziemer's over time like rodents? Therefore, I decided to use both Multiple Sequence Alignment and Pairwise Sequence alignment to demonstrate this hypoyhesis that humans are primarly the ones that are affected by these PSEN-1 mutations near the active site. And, I used Protein Structure Analysis to see the residues on the gamma secretase enzyme to see which amino acids may have been involved. If we look at the multiple sequence alignment between humans(homo sapians, AAH11729.1), orangutans(Pongo abelii,XP_024087009.1), green monkey's (chlorocebus sabaeus, XP_007985388.1), and Night Monkey's (aotus nancymaae, XP_012293630.1) we can see that there are two active site residues at position 257 and position 385. Although we may not see the positions in the code, placing them in Clustal W2 allows us to see the positions. For all four of these species including humans, the active site residue is on a V amino acid (active site binding at 'DLVAVLCPK') at position 257 and a Y amino acid (active site binding at 'DFIFYSVLV') at position 385. All three of the monkey species have the same active site species as the human sequence, and we can therefore say that these PSEN-1 mutations that affect humans near their active site and lead to FAD is seen in monkey sequences. One article called the "First Primate Model of Familial AD: Marmosets with Presenilin Mutation" saw that monkey's can actually develop plaques and tangles in their brain that can ultimately lead to Alziemer's. This is consistent with my analysis that the monkey sequence matches that of a human FAD patient. Now for the Pairwise Sequence alignment, I wanted to compare a human with FAD with a rodent species that is not known to exhibit this disease. I took a human species (homo sapiens, AAH11729.1) shown as X and a rodent species (Microtus ochrogaster,XP_005343368.1) shown as Y and compared their active sites as well. We know that humans have active site residues at positions 257 and 385 and we can compare the amino acids at that residue to those of a rodent. In this case, it is a vole. If we look at position 257 for humans with an active site with the sequence 'DLVAVLCPK,' the rodent has an active site sequence of 'ISVYDLVAV' which is not the same as humans. At position 385, the human active site sequence is 'DFIFYSVLV' while the rodent sequence is 'LGLGDFIFY.' Therefore, the active sites do have some sequences that overlap but the mutations may not affect the same sites since the active site residue is not the same. And we can use this analysis to see that these rodents do not have the same site where the transcription factors can bind to the active site. Therefore these PSEN-1 mutations do not seem to bind to the rodents enzyme to cause this cleavage of amyloid-proteins in order to cause FAD. This is consistent with the idea that mice and other rodents are not known to develop FAD like humans. Additionally, the Protein Structure Analysis was able to demonstrate the enzyme that is affected by these PSEN-1 mutations (Gamma Secretase) and all the atoms/residues within this structure. We can see that PSEN-1 does affect the gamma secretase enzyme in both humans and other species such as Monkey's.
